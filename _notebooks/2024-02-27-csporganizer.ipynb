Mental Health Game Brainstorm

LINK TO OUR CANVA: 

Ideal: Create a site that consists of three games/simulations. First, a questionnaire that helps determine the user’s mental condition. Secondly, a game for the user to play around with, while probably engaging in activities that relate to mental health. Lastly, a report personalized to the user, using API. 

3 questions questionnaire before the game
2 minute simulation game for diagnostic

Depression
High functioning anxiety
Low functioning anxiety

Ex: A game with stimulation that can indicate if the player gets easily distracted or loses attention. Also forces quick decision-making and can indicate symptoms of mental health issues. Other factors could be large cliffs etc to see how the player reacts. We would also add NPC’s and try to make the character report back to them indicating how relatable the game felt.

Step one: Make a questionnaire
By having a simple quiz at the beginning consisting of three questions we can indicate whether a player is portraying symptoms of depression or anxiety. The play may select either the “I might have high-functioning anxiety,” “I might have low-functioning anxiety”, or “I may have depression” quizzes. These will all have three questions. An example would be the depression quiz. This would ask questions like “How often do you feel exhausted or tired, do you have difficulty waking up in the morning?” “do you often feel sad or hopeless?” and “Has anyone had depression in your family before?” By answering either always or never to the first question,  often for the second, and yes for the third, it gives a clear indication that the player could have depression.

Step two: indicate how the player makes decisions 
players go through a series of events that require quick thinking so the player must move quickly. This indicates cognitive impairment, a common problem tied to chemical imbalances in your brain oftentimes causing depression and anxiety. Along 	with this by overstimulating the player so they can get easily distracted we provide an easy way for the player to get off track. If they fall off track then we can indicate that there is a likelihood of mental health issues since loss of attention is a common symptom of mental health issues.

Step three: Collect the data and provide an analysis
After the player has finished the two-minute game along with the questionnaire we can use an API to collect the information given. This will provide a report for the player and tell them how likely they are to have mental health issues. It would then have an encouraging statement as it tells the player to seek out help for a professional diagnosis from a doctor. By using an API to connect to Google or Yahoo we could also locate therapy either online or near where the player lives. 


	Possible games we can code:
Run
Instead of making it slow paced we would have lots of colors and make it fast to stimulate the player. The character would also not die if they went off course but instead would just not get points for running off the course. This would encourage the player to stay on course but also not force them, this would allow for us to determine attention span. Random objects would also pop up giving the player about 2-3 seconds to swerve around them. The inability to do so would indicate the player has slow decision-making.
